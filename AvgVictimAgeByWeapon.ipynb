# visualization EX 1 

```python
import pandas as pd
```

```python
df = pd.read_excel("homicide-reports.xlsx")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
AvgVictimAge_Weapon=df.groupby(df.Weapon)['Victim Age'].mean()
```

```python
AvgVictimAge_Weapon.plot.barh(title='Avg Victim Age By Weapon',figsize=(10,5))
```
